```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
```

```python
dataset,classes = make_blobs(n_samples = 250, n_features=2,centers=1, cluster_std=0.3, random_state=1)
```

```python
df= pd.DataFrame(dataset, columns=['var1', 'var2'])
```

```python
plt.scatter(df['var1'],df['var2'])
```

```python
df.head()
```

```python
classes
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps= 0.27, min_samples=25)
```

```python
pred = dbscan.fit_predict(df)
```

```python
pred
```

plotting the data points of the cluster and the show outliers

```python
outlier_index = np.where(pred== -1)
outlier_value = dataset[outlier_index]
```

```python
outlier_value
```

```python
plt.scatter(df['var1'],df['var2'])
plt.scatter(outlier_value[:,0], outlier_value[:,1], color= 'r')
```

```python
from sklearn.datasets import make_circles
```

```python
datasets, classes = make_circles(n_samples=500, factor=0.3, noise = 0.1)
```

```python
plt.scatter(datasets[:,0],datasets[:,1], c= classes)
```

```python
dbscan = DBSCAN(eps= 0.2, min_samples=3)
```

```python
pred = dbscan.fit_predict(datasets)
```

```python
outlier_index = np.where(pred== -1)
outlier_value = datasets[outlier_index]
```

```python
plt.scatter(datasets[:,0],datasets[:,1],c= classes)
plt.scatter(outlier_value[:,0], outlier_value[:,1], color= 'r')
```

```python
from sklearn.datasets import make_moons
```

```python
dataset, classes = make_moons(n_samples = 500, random_state = 1, noise= 0.15)
```

```python
plt.scatter(dataset[:,0],dataset[:,1],c= classes)
```

```python
dbscan = DBSCAN(eps= 0.2, min_samples=15)

pred = dbscan.fit_predict(dataset)

outlier_index = np.where(pred== -1)
outlier_value = dataset[outlier_index]

plt.scatter(dataset[:,0],dataset[:,1],c= classes)
plt.scatter(outlier_value[:,0], outlier_value[:,1], color= 'r')
```
